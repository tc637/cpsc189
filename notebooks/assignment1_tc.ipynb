```python
import os, stat

def index(directory):
    """
    input:  directory -- string of directory you want to traverse
    output: files     -- list of strings containing all file names and sizes, as well as the
                         total size of directory
    """
    
    # like os.listdir, but traverses directory trees
    first_dir = directory # remember the original directory
    stack = [directory]   # create a list containing the input directory
    files = []            # create an empty list for the output: files (list of strings containing names and file sizes)
    sizes = []            # create an empty list as an accumulator to output total size of directory
   
    while stack:          # while the stack isn't empty
        
        directory = stack.pop()                      # pop the last element out of stack (i.e. last directory in the list)
        
        for file in os.listdir(directory):           # loop over all files in the directory
            
            fullname = os.path.join(directory, file) # get the full name of the file
            st = os.stat(fullname)                   # get the stats of the file
            size = st.st_size                        # extract the size of the file from st
            
            file_str = fullname + ' ' + str(size)    # concatenate file name and size into a single string
            
            files.append(file_str)                   # append the whole string as an element into the files list
            sizes.append(size)                       # append the size of the file to the accumulator
            
            if os.path.isdir(fullname) and not os.path.islink(fullname): # if the 'file' name turns out to be a directory
                stack.append(fullname)                                   # append it to the stack 
    
    total_size = sum(sizes)  # calculate the total directory size
    last_str = "The total size of {} is {} bytes.".format(first_dir,total_size) # create a string to display the total size
    files.append(last_str)   # append that string to the end of files
    
    return files             # output the list of strings


for file in index("C:\Users\Tim\Documents\Github"):
    print(file)
    
# size of my 'Github' directory matches what I've got in Explorer
```
